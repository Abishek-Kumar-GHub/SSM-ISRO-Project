```python
import pandas as pd
from astropy.io import fits

df_name_fits = [
    "HK_file", 
    "TEMPORAL_file",
      "LBT_file", 
      "MKF_file", 
      "ORB_file", 
    "LEVEL_1_fits_file", 
    "Level_2_Clean_SSM_1_Aug_fits_file", 
    "Level_2_Clean_SSM_2_Aug_fits_file", 
    "Level_2_Clean_SSM_3_Aug_fits_file",
]
df_name_prop = [
    "PROP_file_SSM_1", 
    "PROP_file_SSM_2", 
    "PROP_file_SSM_3",
]
file_paths_prop = [
    # SSM 1 prop
    "E://isro//Ssm//20151017_P01_126T03_9000000032_level2//SSM//SSM_00285//AS1P01_126T03_9000000032_00285ssm_prop_ssm1//SSM1_15_1_00285.prop",
    # SSM 2 prop
    "E://isro//Ssm//20151017_P01_126T03_9000000032_level2//SSM//SSM_00285//AS1P01_126T03_9000000032_00285ssm_prop_ssm1//SSM1_15_1_00285.prop",
    # SSM 3 prop
    "E://isro//Ssm//20151017_P01_126T03_9000000032_level2//SSM//SSM_00285//AS1P01_126T03_9000000032_00285ssm_prop_ssm1//SSM1_15_1_00285.prop",
]
file_paths_fits = [
    # Level 1
    "E://c//SSM-data//20240731_A13_081T01_9000006354//20240731_A13_081T01_9000006354_level1//SSM//SSM_47809//hk//AS1A13_081T01_9000006354_47809ssmhk_level1.fits",
    "E://c//SSM-data//20240731_A13_081T01_9000006354//20240731_A13_081T01_9000006354_level1//SSM//SSM_47809//tem//AS1A13_081T01_9000006354_47809ssmtem_level1.fits",
    "E://c//SSM-data//20240731_A13_081T01_9000006354//20240731_A13_081T01_9000006354_level1//SSM//SSM_47809//AS1A13_081T01_9000006354_47809ssm.lbt",
    "E://c//SSM-data//20240731_A13_081T01_9000006354//20240731_A13_081T01_9000006354_level1//SSM//SSM_47809//AS1A13_081T01_9000006354_47809ssm.mkf",
    "E://c//SSM-data//20240731_A13_081T01_9000006354//20240731_A13_081T01_9000006354_level1//SSM//SSM_47809//AS1A13_081T01_9000006354_47809ssm.orb",
    "E://c//SSM-data//20240731_A13_081T01_9000006354//20240731_A13_081T01_9000006354_level1//SSM//SSM_47809//AS1A13_081T01_9000006354_47809ssm_level1.fits",
    # SSM 1 event FITS
    "E://isro//Ssm//20151017_P01_126T03_9000000032_level2//SSM//SSM_00285//evtlstgtifil//AS1P01_126T03_9000000032_00285ssm_level2_clean_ssm1_aug.fits",
    # SSM 2 event FITS
    "E://isro//Ssm//20151017_P01_126T03_9000000032_level2//SSM//SSM_00285//evtlstgtifil//AS1P01_126T03_9000000032_00285ssm_level2_clean_ssm2_aug.fits",
    # SSM 3 event FITS
    "E://isro//Ssm//20151017_P01_126T03_9000000032_level2//SSM//SSM_00285//evtlstgtifil//AS1P01_126T03_9000000032_00285ssm_level2_clean_ssm3_aug.fits",
]

# Function to create DataFrame for FITS files
def create_fits_df(file_paths_fits, df_name_fits):
    dataframes = {}
    for file_path, df_name in zip(file_paths_fits, df_name_fits):
        # Choose the HUD index
        HUD_no = 17 if file_paths_fits.index(file_path) > 6 else 1
        with fits.open(file_path) as hdul:
            # Get column names from the specified HDU
            columns = hdul[HUD_no].columns.names
        # Create a DataFrame with column names
        df = pd.DataFrame({col: [] for col in columns})
        # Store the DataFrame in the dictionary with the file name as the key
        dataframes[df_name] = df
    return dataframes

# Function to create empty DataFrame for property files
def create_empty_dataframe(file_path):
    # Set to store unique column names
    columns = list()

    # Read the file and extract column names
    with open(file_path, "r") as file:
        for line in file:
            line = line.strip()
            if not line or line.startswith("#"):  # Skip comments and empty lines
                continue
            if "=" in line:
                key, _ = line.split("=", 1)
                key = key.strip()                
                columns.append(key)

    # Create an empty DataFrame with the unique column names
    df = pd.DataFrame(columns=columns)  # Sort columns for consistency
    return df

# Create DataFrames for FITS files
dataframes = create_fits_df(file_paths_fits, df_name_fits)

# Create DataFrames for property files
for file_path, df_name in zip(file_paths_prop, df_name_prop):
    dataframes[df_name] = create_empty_dataframe(file_path)

# Print DataFrames for verification
for file_name, df in dataframes.items():
    print(f"DataFrame for {file_name}:")
    print(df)
    print("-" * 40)
    print(type(df))
import pickle

# Save the dataframes dictionary to a file
with open("SSM_dict_dataframe.pkl", "wb") as f:
    pickle.dump(dataframes, f)
```
